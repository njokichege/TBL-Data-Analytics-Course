## **Data Library Loading & Import**

This dataset contains responses from a survey about the telecommunications market in India. It includes questions about the region of India the respondent is from, the type of location they live in (urban/rural), the technology they are currently using, and their experiences with various telecommunications services like email, video calls, and downloading speeds.

The dataset comprises 50 entries and 28 columns, each representing a different question from the survey. The columns cover a wide range of topics, from basic demographic information to detailed questions about the user's experience with their telecommunications service.

Now, let's proceed with a more detailed exploratory data analysis (EDA) to understand the dataset better.

## **Exploratory Data Analysis**

### **Survey Duration**

```python
# Convert timestamp column to datetime
data_df['Timestamp'] = pd.to_datetime(data_df['Timestamp'])

# Find minimum and maximum timestamps
min_timestamp = data_df['Timestamp'].min()
max_timestamp = data_df['Timestamp'].max()

# Calculate duration of survey
survey_duration = max_timestamp - min_timestamp

print("Duration of the survey:", survey_duration)
```

```python
# Plot a histogram of survey responses over time
plt.figure(figsize=(7, 3))
plt.hist(data_df['Timestamp'], bins=30, color='skyblue')
plt.xlabel('No. of Survey Days')
plt.ylabel('Number of Responses')
plt.title('Survey Responses Over Time')
plt.xticks(rotation=45)
plt.show()
```

### **Respondents Location**

```python

# Calculate the number of respondents from each region
region_counts = data_df['1.Which region of India are you from?'].value_counts()

# Plotting
plt.figure(figsize=(4, 4))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Distribution of Respondents by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

# Calculate the number of respondents from each type of location
location_counts = data_df['2.Which type of location do you live in?'].value_counts()

# Plotting
plt.figure(figsize=(4, 4))
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Distribution of Respondents by Location Type')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=data_df, x='1.Which region of India are you from?', hue='2.Which type of location do you live in?')
plt.xlabel('Region of India')
plt.ylabel('Number of Respondents')
plt.title('Types of Locations by Region')
plt.xticks(rotation=45)
plt.legend(title='Location Type')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=data_df, x='1.Which region of India are you from?', hue='12.Which network do you use more often?', palette='tab10')
plt.xlabel('Region of India')
plt.ylabel('Number of Respondents')
plt.title('Distribution of Respondents by Region and Preferred Network')
plt.xticks(rotation=45)
plt.legend(title='Network')
plt.show()
```

```python

plt.figure(figsize=(5, 6))
sns.countplot(data=data_df, x='19.Do you think the telecommunications market in India is tilting towards a monopoly?',palette='tab10')
plt.xlabel('Response')
plt.ylabel('Number of Respondents')
plt.title('Responses to the Telecommunications Market Monopoly Question')
plt.show()
```

```python
plt.figure(figsize=(18, 6))

# Plot histogram for download time
plt.subplot(1, 2, 1)
sns.histplot(data=data_df, x='7.How long does it take you to download a song of 4 minutes?', kde=True)
plt.xlabel('Time to Download Song (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Download Time')
```

```python
from wordcloud import WordCloud

responses_combined = ' '.join(data_df['9.What do you usually do with your phone? (Multiple choices)'])

# Create the word cloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(responses_combined)

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud for Phone Usage')
plt.axis('off')  # Turn off axis
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_df, x='12.Which network do you use more often?', y='7.How long does it take you to download a song of 4 minutes?')
plt.xlabel('Preferred Network')
plt.ylabel('Download Time (minutes)')
plt.title('Download Time vs Preferred Network')
plt.xticks(rotation=45)
plt.show()
```

### **Univariate Analysis**

```python
# Visualize the distribution of categorical variables
for column in data_df.select_dtypes(include=['object']):
    sns.countplot(y=column, data=data_df)
    plt.show()

# Visualize the distribution of numerical variables
for column in data_df.select_dtypes(include=['int64', 'float64']):
    sns.histplot(data_df[column])
    plt.show()
```

### **Bivariate Analysis**

```python
# Explore relationships between variables
sns.pairplot(data_df)
plt.show()

# Correlation heatmap for numerical variables
corr_matrix = data_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

* A correlation coefficient (r) close to 1 indicates a strong positive linear relationship between two variables. This means that as one variable increases, the other variable also tends to increase.

* A correlation coefficient close to -1 indicates a strong negative linear relationship between two variables. This means that as one variable increases, the other variable tends to decrease.

* A correlation coefficient close to 0 indicates no linear relationship between the two variables.


So, when we say "strength is towards 1", it means that the correlation coefficient is closer to 1, indicating a stronger positive linear relationship between the variables. Similarly, "strength is towards -1" means that the correlation coefficient is closer to -1, indicating a stronger negative linear relationship between the variables.
