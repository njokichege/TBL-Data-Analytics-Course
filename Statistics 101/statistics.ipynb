#**Statistics 101**

**Introduction:**

Statistics is a powerful tool for understanding and interpreting data. One fundamental concept in statistics is the notion of moments, which provide valuable insights into the characteristics of a data distribution. Moments are mathematical quantities that summarize various aspects of the distribution's shape, central tendency, spread, and symmetry.

The four moments commonly used in statistics are mean, variance, skewness, and kurtosis. Each of these moments captures a different aspect of the data distribution, helping analysts to better understand its properties and make informed decisions.

## **1. Mean (First Moment):**
The mean, often denoted by μ, represents the average value of a dataset. It is calculated by summing all the values in the dataset and dividing by the total number of observations. The mean provides a measure of central tendency, indicating where the "center" of the data is located. It is sensitive to outliers, meaning that extreme values can significantly influence its value.

## 2 ** Variance (Second Moment):**

Variance, denoted by σ², measures the spread or dispersion of the data points around the mean. It quantifies how much the individual data points deviate from the mean. A high variance indicates that the data points are spread out widely, while a low variance suggests that the data points are clustered closely around the mean. Variance is calculated by averaging the squared differences between each data point and the mean.

## **3. Skewness (Third Moment):**

Skewness measures the asymmetry of the distribution. A distribution is considered symmetric if it is balanced around its mean, with equal tails on both sides. Positive skewness indicates that the right tail of the distribution is longer or fatter than the left tail, while negative skewness indicates the opposite. Skewness helps identify whether the data distribution is skewed to the left or right, relative to its mean.

## **4. Kurtosis (Fourth Moment):**

Kurtosis quantifies the degree of peakedness or flatness of the distribution's shape. A high kurtosis indicates that the distribution has heavy tails and a sharp peak, whereas a low kurtosis suggests a flatter distribution with lighter tails. Kurtosis helps distinguish between different types of distributions, such as normal, leptokurtic (heavy-tailed), and platykurtic (light-tailed) distributions.

## **3. Probability Distributions**
Probability distributions describe the likelihood of different outcomes in a random experiment. Common probability distributions include the binomial distribution, Poisson distribution, and normal distribution. Understanding these distributions is crucial for modeling and analyzing data.

```python
from scipy.stats import binom, poisson, norm
import matplotlib.pyplot as plt
```

```python
# Binomial Distribution
n = 10  # Number of trials
p = 0.5  # Probability of success
binomial_dist = binom(n, p)
x = range(n + 1)
plt.bar(x, binomial_dist.pmf(x))
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
# Poisson Distribution
mu = 3  # Mean rate
poisson_dist = poisson(mu)
x = range(10)
plt.bar(x, poisson_dist.pmf(x))
plt.title("Poisson Distributions")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

```python
# Normal Distribution
mean = 0  # Mean
std_dev = 1  # Standard deviation
normal_dist = norm(mean, std_dev)
x = np.linspace(-4, 4, 100)
plt.plot(x, normal_dist.pdf(x))
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```

This example showcases three common probability distributions: binomial, Poisson, and normal distributions. We used Python's SciPy library to visualize the probability mass/function of each distribution. The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. The Poisson distribution models the number of events occurring in a fixed interval of time or space. The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that describes the data clustering around the mean with a specific standard deviation.

```python
import scipy.stats as stats

# Generate sample data (normal distribution)
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate mean
mean = np.mean(data)

# Calculate variance
variance = np.var(data)

# Calculate skewness
skewness = stats.skew(data)

# Calculate kurtosis
kurtosis = stats.kurtosis(data)

# Print the moments
print("Mean:", mean)
print("Variance:", variance)
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)

# Plot histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')

# Plot a vertical line at the mean
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(variance, color='g', linestyle='dashed', linewidth=1, label='Variance')


plt.title('Histogram of Sample Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

This code snippet generates a sample of 1000 data points from a normal distribution with a mean of 0 and standard deviation of 1. It then calculates the mean, variance, skewness, and kurtosis of the sample data using the appropriate functions from numpy and scipy.stats. Finally, it visualizes the data distribution using a histogram, overlaying a dashed red line representing the mean.

You can observe how each moment provides unique insights into the characteristics of the data distribution:

Mean gives us the central tendency of the distribution.
Variance gives us a measure of the spread of the data points around the mean.
Skewness indicates the asymmetry of the distribution.
Kurtosis quantifies the peakedness or flatness of the distribution's shape.

```python
import pandas as pd
import seaborn as sns

# Generate sample data
np.random.seed(0)
num_employees = 500
salaries = np.random.normal(loc=50000, scale=15000, size=num_employees)  # Assume a normal distribution of salaries
attrition_rates = np.random.beta(a=2, b=5, size=num_employees)  # Assume attrition rates follow a beta distribution

# Create a DataFrame
data = pd.DataFrame({'Salary': salaries, 'Attrition Rate': attrition_rates})

# Visualize the distribution
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Salary', y='Attrition Rate', data=data, alpha=0.6)
plt.title('Attrition Rate vs. Salary')
plt.xlabel('Salary')
plt.ylabel('Attrition Rate')
plt.grid(True)
plt.show()

# Calculate skewness
skewness = data['Attrition Rate'].skew()
print("Skewness:", skewness)
```

* We generate a sample dataset where salaries follow a normal distribution and attrition rates follow a beta distribution.
* We create a scatter plot to visualize the relationship between salary and attrition rate.
* We calculate the skewness of the attrition rate distribution.

The skewness value will indicate whether the distribution of attrition rates is skewed towards higher or lower salary levels. Positive skewness suggests a longer tail on the right side (higher salaries), while negative skewness suggests a longer tail on the left side (lower salaries). This can provide insights into whether salary plays a role in attrition rates.

```python
# Generate sample data
np.random.seed(0)
num_users = 1000
negative_feedback = np.random.randint(1, 11, size=num_users)  # Assuming negative feedback scores on a scale of 1 to 10
attrition = np.random.choice([0, 1], size=num_users, p=[0.8, 0.2])  # Binary attrition indicator (0: stay, 1: leave)

# Create a DataFrame
data = pd.DataFrame({'Negative Feedback': negative_feedback, 'Attrition': attrition})

# Calculate mean, median, and mode of negative feedback
mean_feedback = data['Negative Feedback'].mean()
median_feedback = data['Negative Feedback'].median()
mode_feedback = data['Negative Feedback'].mode().values[0]

# Plotting
plt.figure(figsize=(12, 8))

# Histogram for negative feedback distribution
sns.histplot(data['Negative Feedback'], bins=10, kde=True, color='skyblue', edgecolor='black', linewidth=1.2, label='Distribution')

# Highlight mean, median, and mode
plt.axvline(mean_feedback, color='red', linestyle='--', label='Mean')
plt.axvline(median_feedback, color='green', linestyle='--', label='Median')
plt.axvline(mode_feedback, color='orange', linestyle='--', label='Mode')

print("Mean: ", mean_feedback)
print("Meadian: ",median_feedback)
print("Mode: ", mode_feedback)

print("Skewness:", data['Negative Feedback'].skew())
# Add legend and labels
plt.legend()
plt.title('Negative Feedback Distribution and Attrition')
plt.xlabel('Negative Feedback Score')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

* We generate a sample dataset containing negative feedback scores (on a scale of 1 to 10) and a binary indicator for attrition (0: stay, 1: leave).
* We calculate the mean, median, and mode of the negative feedback scores.
* We plot a histogram to visualize the distribution of negative feedback scores, while also highlighting the mean, median, and mode with dashed lines.

This plot will allow you to see how the distribution of negative feedback scores relates to user attrition from the platform, as well as the central tendency and typical values of negative feedback. It provides a comprehensive view of the relationship between negative feedback and user attrition.
